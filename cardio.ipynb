```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Importer les données
data = pd.read_csv("CardioGoodFitness.csv")

# Sélectionner uniquement les colonnes numériques
numeric_data = data.select_dtypes(include='number')

# Standardisation des données
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# Calculer la matrice de corrélation
corr = np.corrcoef(scaled_data, rowvar=False)

# Calculer les valeurs et les vecteurs propres
eig_vals, eig_vecs = np.linalg.eig(corr)

# Calculer le pourcentage de variance expliquée par chacun des axes factoriels
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

# Afficher les résultats
result_df = pd.DataFrame({'Valeur propre': eig_vals, 'Variance expliquée (%)': var_exp, 'Variance cumulée (%)': cum_var_exp})
print(result_df)
# Calculer les valeurs et les vecteurs propres
eig_vals, eig_vecs = np.linalg.eig(corr)

# Afficher les résultats
print("Valeurs propres:")
print(eig_vals)
print("\nVecteurs propres:")
print(eig_vecs)

# Plot des valeurs propres
plt.bar(range(1, len(eig_vals) + 1), eig_vals)
plt.xlabel('Composante principale')
plt.ylabel('Valeur propre')
plt.title('Plot des valeurs propres')
plt.show()

# Plot de la variance expliquée
plt.plot(range(1, len(var_exp) + 1), var_exp, marker='o')
plt.xlabel('Composante principale')
plt.ylabel('Variance expliquée')
plt.title('Plot de la variance expliquée')
plt.show()
```

**Classification Ascendante Hiérarchique (CAH)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Importer les données
data = pd.read_csv("CardioGoodFitness.csv")

# Sélectionner uniquement les colonnes numériques
numeric_data = data.select_dtypes(include='number')

# Standardisation des données
scaled_data = (numeric_data - numeric_data.mean()) / numeric_data.std()

# Calculer la matrice de distance
Z = linkage(scaled_data, method='ward', metric='euclidean')

# Afficher le dendrogramme
plt.figure(figsize=(12, 8))
dendrogram(Z)
plt.title('Dendrogramme de la CAH')
plt.xlabel('Individus')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Initialisez une liste pour stocker les valeurs de coefficient de silhouette
silhouette_scores = []

# Nombre maximal de classes à tester
max_clusters = 10

# Itérer sur différents nombres de classes
for n_clusters in range(2, max_clusters + 1):
    # Instanciez le modèle de CAH
    cah_model = AgglomerativeClustering(n_clusters=n_clusters)

    # Effectuez la CAH sur les données
    clusters = cah_model.fit_predict(scaled_data)

    # Calculez le coefficient de silhouette
    silhouette_avg = silhouette_score(scaled_data, clusters)

    # Ajoutez le coefficient de silhouette à la liste
    silhouette_scores.append(silhouette_avg)

# Trouvez le nombre de classes qui maximise le coefficient de silhouette
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + 2  # +2 car l'index commence à 0 et nous avons commencé à partir de 2 classes

print("Nombre optimal de classes selon la méthode de la silhouette :", optimal_num_clusters)
```

```python
# Instancier le modèle de CAH avec le nombre optimal de classes retenu
cah_model = AgglomerativeClustering(n_clusters=optimal_num_clusters)

# Effectuer la CAH sur les données
clusters = cah_model.fit_predict(scaled_data)

# Visualisation graphique de la partition obtenue sur les plans factoriels de projection
plt.figure(figsize=(10, 6))
for cluster in range(optimal_num_clusters):
    plt.scatter(scaled_data.values[clusters == cluster, 0], scaled_data.values[clusters == cluster, 1], label=f'Cluster {cluster+1}')
plt.title('Partition obtenue sur les plans factoriels de projection')
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV
cardio_data = pd.read_csv("CardioGoodFitness.csv")

# Sélectionner uniquement les colonnes numériques
numeric_data = cardio_data.select_dtypes(include='number')

# Calcul de l'inertie pour différents nombres de clusters
inertias = []
for i in range(1, 11):  # Tester pour un nombre de clusters de 1 à 10
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(numeric_data)
    inertias.append(kmeans.inertia_)

# Tracé de la courbe du coude
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel("Inertie")
plt.title('Méthode du Coude pour le Choix du Nombre de Clusters')
plt.show()
```

**Interprétation des classes : statistiques comparatives **

```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np

# Charger les données depuis le fichier CSV
cardio_data = pd.read_csv("CardioGoodFitness.csv")

# Sélectionner uniquement les colonnes numériques
numeric_data = cardio_data.select_dtypes(include='number')

# Instanciation du StandardScaler
scaler = StandardScaler()

# Transformation – centrage-réduction
cardio_cr = scaler.fit_transform(numeric_data)

# Application de l'algorithme K-means
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(cardio_cr)

# Ajout des informations de clusters aux données
cardio_data['cluster'] = clusters

# Calcul des moyennes des caractéristiques de chaque cluster
cluster_means = cardio_data.select_dtypes(include='number').groupby('cluster').mean()
print("Moyennes des caractéristiques de chaque cluster :")
print(cluster_means)


# Calcul des différences significatives entre les clusters
cluster_diff = cluster_means.diff().iloc[1:]  # Différences par rapport au premier cluster
print("\nDifférences significatives entre les clusters :")
print(cluster_diff)

# Caractéristiques les plus discriminantes pour chaque cluster
print("\nCaractéristiques les plus discriminantes pour chaque cluster :")
for i, row in cluster_diff.iterrows():
    print(f"Cluster {i}:")
    max_diff_feature = row.idxmax()
    max_diff_value = row.max()
    print(f"Caractéristique la plus discriminante: {max_diff_feature}, Différence: {max_diff_value}")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Appliquer l'ACP sur les données standardisées
acp = PCA(n_components=2).fit_transform(cardio_cr)

# Projeter dans le plan factoriel avec un code couleur différent selon le groupe
# Remarquer le rôle de zip() dans la boucle
for couleur, k in zip(['red', 'blue', 'lawngreen', 'aqua'], range(4)):
    plt.scatter(acp[clusters == k, 0], acp[clusters == k, 1], c=couleur)

plt.xlabel('Première composante principale')
plt.ylabel('Deuxième composante principale')
plt.title('Projection des données dans le plan factoriel')
plt.show()
```
